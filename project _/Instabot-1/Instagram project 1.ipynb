# 1)login Instagram Handle

```python

time.sleep(2)
def login(user_name_1,password_1):
    wait=WebDriverWait(driver,10)
    user_name=wait.until(EC.presence_of_element_located( (By.NAME, 'username') ))
    user_name.send_keys(user_name_1)
    password=driver.find_element_by_name('password')
    password.send_keys(password_1)
    password.submit()
login(user_name_1,password_1)
```

# 5)Like/Unlike posts
#### 1)Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
def like_posts(handle):
    get_profile(handle)
    # web page scrool
    current_height = driver.execute_script('return document.body.scrollHeight;')
    wait = WebDriverWait(driver,20)
    while True:
        driver.execute_script('window.scrollTo(0,arguments[0]);' , current_height)
        time.sleep(3) # wait for 3 seconds to load new posts
        posts = driver.find_elements_by_xpath('//div[@class = "_9AhH0"]') # all posts on page
        if len(posts)>=30:
            break
        current_height = driver.execute_script('return document.body.scrollHeight;')
    for i in range(30):
        posts[i].click()
        like_btn=wait.until(EC.presence_of_element_located( (By.XPATH,'//section[contains(@class , "ltpMr")]/span[1]/button') ))
        data = BeautifulSoup(driver.page_source,'lxml')
        section=data.find(class_="ltpMr Slqrh")
        butn=section.find('button')
        like=butn.svg.get('aria-label').strip()
        if like=='Like':
            like_btn.click()
        else:
            print('Once Liked')
        close_post_btn=driver.find_element_by_xpath('//div[contains(@class,"_2dDPU")]/div[3]/button')
        close_post_btn.click()
        time.sleep(5)
like_posts('dilsefoodie')
```

# 6)Extract list of followers
1) Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def extract_follower(handle):
    get_profile(handle)
    follower=driver.find_elements_by_class_name('-nal3 ')
    time.sleep(3)
    follower=follower[1]
    follower.click()
    time.sleep(5)
    while True:
            # locate any element on pop_up page
            obj_created_for_pop_up_scroll = driver.find_element_by_xpath('//a[contains(@class,"FPmhX notranslate")]')
            followers_list = driver.find_elements_by_class_name('wFPL8')
            if len(followers_list) >= 500:
                break
            obj_created_for_pop_up_scroll.send_keys(Keys.END)
            time.sleep(2) # wait to load new followers

    for i in range(500):
        print(str(i+1)+") ",followers_list[i].get_attribute('innerHTML'))

extract_follower('sodelhi')
```

# Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
If You have already seen the story.

Or The user has no story.

Or View the story if not yet seen.

```python
get_profile('coding.ninjas')
```

```python
wait = WebDriverWait(driver,10)
def check_story(handle_name):
    get_profile(handle_name)
    time.sleep(2)
    try:
        story = wait.until(EC.presence_of_element_located( (By.CLASS_NAME,'RR-M-.h5uC0') ))
        
        # If story is not seen, then height is 210, After viewing story height becomes 208
        height = driver.find_element_by_class_name('CfWVH').get_attribute('height')
        if int(height)==208:
            print("Already seen")
        else:
            print("Viewing the story...")    
            driver.execute_script('arguments[0].click();',story)
    except:
        print(handle_name,"has no story")
    
check_story('coding.ninjas')
```
