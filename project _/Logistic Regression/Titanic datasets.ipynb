```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
```

#### Loading Training data

```python
df_train=pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")
df_train.head()
```

#### Loading Testing data

```python
df_test=pd.read_csv("0000000000002429_test_titanic_x_test.csv")
df_test.head()
```

#### Removing Cabin from Test and training data

```python
df_test.drop('Cabin',axis=1,inplace=True)
df_train.drop('Cabin',axis=1,inplace=True)
```

### In this section converting sex Male and Female in '1', '0'.

```python
df_test['Sex'].replace('male',1,inplace=True)
df_test['Sex'].replace('female',0,inplace=True)   
df_train['Sex'].replace('male',1,inplace=True)
df_train['Sex'].replace('female',0,inplace=True)   
```

#### In Section This We Will Nan Value Of Age.

```python
train_age_mean=df_train['Age'].mean()
df_train['Age'].fillna(train_age_mean,inplace=True)## Filling NAN  data in age column in with mean value in training data se
testing_age_mean=df_test['Age'].mean()
df_test['Age'].fillna(testing_age_mean,inplace=True)## Filling NAN  data in age column in with mean value in training data se
```

#### In this section we will string present in embarked 'S','C','Q' with  0,1.

```python
df_test['Embarked'].replace('C',1,inplace=True)
df_test['Embarked'].replace('S',0,inplace=True)   
df_train['Embarked'].replace('C',1,inplace=True)
df_train['Embarked'].replace('S',0,inplace=True)   
df_test['Embarked'].replace('Q',2,inplace=True)   
df_train['Embarked'].fillna(1,inplace=True)
df_train['Embarked'].replace('Q',2,inplace=True)
df_test['Embarked'].fillna(1,inplace=True)
df_train.head(20)
```

#### Drop Unwanted column

```python
df_test.drop(['Ticket' ,'Name'],axis=1,inplace=True)
df_train.drop(['Ticket' ,'Name'],axis=1,inplace=True)
df_test.head(20)
```

#### Converting in training and testing NUmpy aarray

```python
y_train=np.array(df_train['Survived'])
df_train.drop('Survived',axis=1,inplace=True)
x_train=df_train.to_numpy(copy=True)
x_test=df_test.to_numpy(copy=True)
```

```python
clf = LogisticRegression(solver='liblinear')
clf.fit(x_train,y_train)
```

```python
y_predictions = clf.predict(x_test)
```

```python
y_predictions = y_predictions.reshape(-1,1)
print(y_predictions.shape)
```
